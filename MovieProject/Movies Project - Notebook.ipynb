### Data Analysis 


Checking the distribution of `duration` column using a histogram.

Checking how many movies have the `imdb_score` less than `4`.

### Further Analysis
    
#### Finding in which year we have the `highest` and `lowest` number of movies

Finding out number of movies per year

```python
count_per_year = data.groupby("title_year").size()
count_per_year
```

Creating a line chart to show the result

```python
ax = count_per_year.plot(figsize = (8,6))
ax.set_xlabel("Years")
ax.set_ylabel("Movie Count")
ax.set_title("Distribution of Movies Through the Years")
```

#### Finding in which year we have `minimum` and `maximum` average `imdb_score` 

Finding average imdb score per year 

```python
avg = data.groupby("title_year")['imdb_score'].mean()
avg
```

Using a barchart to visualize the result

```python
ax = avg.plot.bar(figsize = (18,8), title = "Average IMDB score per Year")
ax.set_xlabel("Years")
ax.set_ylabel("Score")
```

Finding In which year we have the `minimum` and `maximum` spent movie budget? 

```python
ax = data.groupby("title_year")["movie_budget"].sum().plot()
ax.set_xlabel("Year")
ax.set_ylabel("Budget")
ax.set_title("Movie Budget through the Years")
```

Correlation between IMdb score of a movie and the movie income

To approach this

```python
data_corr = data[["imdb_score", "movie_income"]]
data_corr
```
